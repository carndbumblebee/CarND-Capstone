```python
def tl_light_classifier(image):
    width_to_trim = int(image.shape[1] / 3)
    # height_to_trim = int(image.shape[0] / 10)
    height_to_trim = 0
    sub_image = image[height_to_trim:(image.shape[0]-height_to_trim), width_to_trim:(image.shape[1]-width_to_trim)]
    
    G_threshold = (200, 255)
    R_threshold = (200, 255)
    B_threshold = (200, 255)
    
    R = sub_image[:,:,0]
    G = sub_image[:,:,1]
    B = sub_image[:,:,2]
    
    B_binary = np.zeros_like(B)
    B_binary[(B > B_threshold[0]) & (B <= B_threshold[1])] = 1
    
    G_binary = np.zeros_like(G)
    G_binary[(G > G_threshold[0]) & (G <= G_threshold[1])] = 1
    
    R_binary = np.zeros_like(R)
    R_binary[(R > R_threshold[0]) & (R <= R_threshold[1])] = 1
    
    threshold = 0.08
    isRed = (np.sum(R_binary) / R_binary.size) > threshold
    isGreen = (np.sum(G_binary) / G_binary.size) > threshold
    isBlue = (np.sum(B_binary) / B_binary.size) > threshold
    
    top_image = G_binary[0:int(G_binary.shape[0]/2), :]
    bottom_image = G_binary[int(G_binary.shape[0]/2):G_binary.shape[0], :]

    if np.sum(bottom_image) == 0:
        is_edge = True
    elif np.sum(top_image) / np.sum(bottom_image) > 5:
        is_edge = True
    else:
        is_edge = False
    
    if isRed and isGreen and isBlue:
        return 'Green'
    elif isRed and isGreen and is_edge:
        return 'Red'
    elif isRed and isGreen:
        return 'Yellow'
    elif isRed:
        return 'Red'
    elif isGreen:
        return 'Green'
    else:
        return 'Uknown'
```

```python
for filename in glob.glob('../tl_test_images/*.jpg'):
    image = cv2.imread(filename)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    plt.figure
    plt.imshow(image)
    plt.show()
    
    print(filename)
    print(tl_light_classifier(image))
```
